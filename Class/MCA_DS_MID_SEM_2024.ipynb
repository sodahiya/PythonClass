**1. Write a Python Program to accept 10 numbers from the user and store them in a list,find the maxiumm number in the above list.**


```python
number=[None] *10
numCount = 0;
while (numCount<10):
    try :
      number[numCount]=int(input("Enter " + str(numCount+1) +  " number in the list :"))
    except:
      print("Enter a valid number!") #if invalid number entered
      continue
    numCount+= 1 #increment in case of valid number only
print(number)
max= number[0] #init max to the first number
for i in  range(9): #compare with the remaining 9 numbers
    if number[i+1]>max:
        max=number[i+1]

print("Maxiumm Number In The List",max)
```

**2.A program for accept number of days from the user.then convert it to number of years,months,week,days.using try except.**

```python
while(True):
  try:
    days=int(input("Enter Number of Days:")) # Input number of Days
    break #break the loop if valid number entered
  except:
    print("Enter a valid number!") #if invalid number entered
year=days//365 #calculate year
remaining_days=days%365 #calculate remaining days
month=remaining_days//30 #calculate month
month_remaining_days=remaining_days%30 #calculate remaining days after getting month
week=month_remaining_days//7 #calculate Week
final_days=month_remaining_days%7 #Left Days after fetching the week.
print(days, "days = ", year, " year(s), ", month, " month(s), ", week, " week(s), and ", final_days, " day(s).")
```

**2.b A python function that prints the reverse of a given number.**

```python

def reverse_number(number): #function to reverse the number
    reverse=0
    while(number > 0): #loop continue untill the number is greater than zero.
        digit=number%10 #calculating last digit of number
        reverse=reverse*10+digit #adding last digit to reverse
        number=number//10 #removing last digit from number
    return reverse #return the reverse number

value=23456789 #intialise the number with that
output=reverse_number(value) #call the function
print("Reverse number of ", value, " is ", output) #print the output
```

**3.A program that accept the filename from the user and the mode in which the file should be opened.whether it should be opened in append mode or write mode.then it keeps accepting input from the user and writing them to the file.finally when the user type <Done>,the file is closed**

```python
try:
  filename = input("Enter the file name: ") #input the file name

  while(True): #loop for valid input
    mode = input("Enter the mode (a for append, w for write): ")


    if mode not in ['a', 'w']: #check the mode is valid or not
        print("Invalid mode. Please enter 'a' or 'w'.")
        continue
    else: #if valid mode break the loop
      break

  with open(filename, mode) as fout: #open the file in append or write mode
    line = input("Enter input to write to the file (enter '<Done>' to finish): ")
    while line != '<Done>': #loop untill user enter <Done>
      fout.write(line + '\n') #writing in the file
      line = input("Enter input to write to the file (enter '<Done>' to finish): ") #input again
    else: #if user enter <Done> then close the file
      fout.close() #close the file
      print("File closed.")

except FileNotFoundError:
  print("File not found.")

except PermissionError:
  print("Permission denied to access the file.")

except Exception as e:
  print("An error occurred: ", e)


```
